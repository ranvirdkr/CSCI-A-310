**Problem 2.** The times variable is set to 1,000,000 which determines how many times the dice game will be simulated. And money = 0 is initialized to 0 keeping a running total of the total money gained or lost. The for loop will run all the iterations, simulating one game of dice in each iteration. d1 and d2 are inside the loop, and these lines simulate rolling two standard six-sided dice and the random.randint(1,6) generates a random integer between 1 and 6 for each die. sum=d1+d2 is the sum of 2 dice and stores in this variable. if sum==7: is a condition to check if the sum of the dice is exactly 7 and it gives the player +4 units of money else: -1 unites which are both taken from the money. Once the loop is finished with all 1,000,000 games it will print money/timnes to calculate the average money won or lost pergame.

```python
import random

times = 1000000
money = 0

for i in range(times):
  d1 = random. randint(1, 6)
  d2 = random. randint(1, 6)
  sum = d1 + d2
  if sum == 7:
    money += 4
  else:
    money-= 1

print (money/times)
```

**Problem 3.** This results in the 100 lockers whether they are closed or opened. The ones that are opened are the perfect squares (1,4,9,16,25,36,49,64,81,100) because of the different factors with for example student 1 opening locker 1 and locker 4 is opened by students 1,2,and 4.

```python
lockers =[' '] * 100
for student in range(100):
  for locker in range(len(lockers)):
    if (locker+1) % (student+1)==0:
      if lockers[locker]==' ': #locker is closed
        lockers[locker]='*'  #open it
      else:
        lockers[locker]=' '  #otherwise close the locker
  print(''.join(lockers))
```
