```python
class LinkedList:
  def __init__(self, num, next):
    self.num = num
    self.next = next
  def insert(self, num):
    if self.next == None:
      self.next = LinkedList(num, None)
    else:
      self.next.insert(num)
a = LinkedList(3, LinkedList(1, LinkedList(5, None)))
a.insert(2)
```
